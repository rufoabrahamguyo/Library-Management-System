```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data
y = iris.target
X_train,x_test,y_train,y_test = train_test_split(X,y,random_state=0)

tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)

#visualize the model in form of tree
from sklearn.tree import export_graphviz
export_graphviz(tree,out_file="tree.dot",class_names=["setosa","versicolor","virginica"],feature_names=iris.feature_names,impurity=False,filled=True,rounded=True)
```

```python
import graphviz
with open ("tree.dot")as f:
  tree_graph=f.read()
graphviz.Source(tree_graph)
```

```python
from sklearn import tree
tree.plot_tree(tree_model,class_names=["setosa","versicolor","virginica"],feature_names=iris.feature_names,impurity=False,filled=True,rounded=True)
```

```python
print("Accuarcy of training set {:.2f}".format(tree_model.score(X_train,y_train)))
print("Accuarcy of test set {:.2f}".format(tree_model.score(x_test,y_test)))
```
