```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target,random_state=42)
logreg=LogisticRegression(max_iter=100000).fit(X_train, y_train)
print("Train set score: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
logreg100=LogisticRegression(C=100,max_iter=100000).fit(X_train, y_train)
print("Train set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test, y_test)))
```

```python
logreg001=LogisticRegression(C=0.001,max_iter=100000).fit(X_train, y_train)
print("Train set score: {:.3f}".format(logreg001.score(X_train, y_train)))
print("Test set score: {:.3f}".format(logreg001.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure coefficient arrays are 1D for plotting
plt.plot(np.abs(logreg.coef_)[0], 'o', label="C=1")
plt.plot(np.abs(logreg100.coef_)[0], '^', label="C=100")
plt.plot(np.abs(logreg001.coef_)[0], 'v', label="C=0.001")

# Proper x-tick labels
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])

# Legend and labels
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-5, 5)
plt.xlabel("Feature index")
plt.ylabel("Coefficient magnitude (absolute)")
plt.title("Effect of C on Logistic Regression Coefficients")
plt.tight_layout()
plt.legend()
plt.show()

```
